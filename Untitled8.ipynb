```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv('data.csv', na_values=['', ' '])

print(df.head())


threshold = 0.5 * len(df)
df = df.dropna(axis=1, thresh=threshold)
print(df.info())

```

**Проверяем на наличие дубликатов**

```python
pd.options.display.max_columns = 500
df.head()
```

```python
duplicates_count = df.duplicated().sum()

print(f"\nКоличество дубликатов: {duplicates_count}")
```

```python
# Создадим целевую переменную
df['Срок поиска артефакта'] = (pd.to_datetime(df['Дата окончания поиска']) - pd.to_datetime(df['Дата начала поиска'])).dt.total_seconds() / (3600 * 24)


# Удалим столбцы с датами
df = df.drop(['Дата начала поиска', 'Дата окончания поиска'], axis=1)
# Удалим столбец Общее время поиска т.к. он дублирует целевую переменную
df = df.drop(['Общее время поиска', 'Общее время активного поиска', 'Общее время простоев'], axis=1)

numeric_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = df.select_dtypes(include=['object']).columns.tolist()

print("\nЧисловые признаки:", numeric_features)
print("Категориальные признаки:", categorical_features)
```

```python
for col in numeric_features:
    # Проверка наличия отрицательных значений
    negative_values = df[col] < 0
    if negative_values.any():
        print(f"Найдено отрицательных значений в столбце '{col}': {negative_values.sum()}")
        # Замена отрицательных значений на NaN
        df.loc[negative_values, col] = np.nan
    else:
        print(f"Отрицательных значений в столбце '{col}' нет")
```

Заполним оставшиеся пропуски

Категориальные признаки заполним значением 'Not given',
числовые признаки заполним средним



```python
for col in numeric_features:
    df[col].fillna(df[col].mean(), inplace=True)

for col in categorical_features:
    df[col].fillna('not given', inplace=True)

# Проверка наличия пропусков
print("\nПропущенные значения после обработки:")
print(df.isnull().sum())
```

```python
for col in categorical_features:
    unique = df[col].nunique()
    print(f"{col}: {unique} уникальных значений")
```

Делаем вывод, что OneHotEncoding для всех категориальных признаков не подойдет

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Срок поиска артефакта'], kde=True)
plt.title('Срок Поиска Артефакта')
plt.xlabel('Срок поиска')
plt.ylabel('Частота')
plt.show()
```

Так как график сильно скошен, мы видим что

1. Большинство артефактов находятся в довольно короткий срок
2. Присутсвуют выбросы, малая часть поисков занимет >100 дней



```python
# Удаление выбросов 
q_high = df['Срок поиска артефакта'].quantile(0.95)
df = df[df['Срок поиска артефакта'] <= q_high]
```

Масштабируем данные

```python
from sklearn.preprocessing import StandardScaler

numeric_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_features.remove('Срок поиска артефакта')  # Удаляем целевую переменную из списка

scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

# Обработка категориальных признаков

Для признаков с маленьким количеством категорий используем OneHotEncoding, а для остальных FrequencyEncoding

```python
low_categority = [col for col in categorical_features if df[col].nunique() <= 10]
high_categority = [col for col in categorical_features if df[col].nunique() > 10]

for col in high_categority:
    freq = df[col].value_counts() / len(df)
    df[col] = df[col].map(freq)

df = pd.get_dummies(df, columns=low_categority, drop_first=True)
```

# Разделение на текстовую и обучающую выборки

```python
X = df.drop(['Срок поиска артефакта'], axis=1)
y = df['Срок поиска артефакта']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

# Выбор и обучение модели

RandomForestRegressor

```python
rf = RandomForestRegressor(random_state=1)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)

mae_rf = mean_absolute_error(y_test, rf_preds)
rmse_rf = np.sqrt(mean_squared_error(y_test, rf_preds))
r2_rf = r2_score(y_test, rf_preds)
```

XGBRegressor

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(random_state=1, objective='reg:squarederror')
xgb.fit(X_train, y_train)
xgb_preds = xgb.predict(X_test)

mae_xgb = mean_absolute_error(y_test, xgb_preds)
rmse_xgb = np.sqrt(mean_squared_error(y_test, xgb_preds))
r2_xgb = r2_score(y_test, xgb_preds)
```

Линейная регрессия

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
lr_preds = lr.predict(X_test)

mae_lr = mean_absolute_error(y_test, lr_preds)
rmse_lr = np.sqrt(mean_squared_error(y_test, lr_preds))
r2_lr = r2_score(y_test, lr_preds)
```

Support Vector Regressor

```python
from sklearn.svm import SVR

svr = SVR()
svr.fit(X_train, y_train)
svr_preds = svr.predict(X_test)

mae_svr = mean_absolute_error(y_test, svr_preds)
rmse_svr = np.sqrt(mean_squared_error(y_test, svr_preds))
r2_svr = r2_score(y_test, svr_preds)
```

```python
# Сравнение результатов моделей
results = pd.DataFrame({
    'Модель': ['Random Forest', 'Linear Regression', 'Support Vector Regressor', 'XGBRegressor'],
    'MAE': [mae_rf, mae_lr, mae_svr, mae_xgb],
    'RMSE': [rmse_rf, rmse_lr, rmse_svr, rmse_xgb],
    'R²': [r2_rf, r2_lr, r2_svr, r2_xgb]
})

print("\nСравнение моделей:")
print(results)
```

Random forest больше всего подходит для решения этой задачи


```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
import numpy as np

# Определение диапазона гиперпараметров
param_dist_rf = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Инициализация RandomForestRegressor
rf = RandomForestRegressor(random_state=42)

# Инициализация RandomizedSearchCV
random_search_rf = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist_rf,
    n_iter=5,  # Количество случайных комбинаций
    cv=3,        # Количество фолдов кросс-валидации
    verbose=2,
    random_state=42,
    n_jobs=-1,
    scoring='neg_mean_absolute_error'
)

# Обучение RandomizedSearchCV
random_search_rf.fit(X_train, y_train)

# Лучшие параметры
best_rf = random_search_rf.best_estimator_
print(f"Лучшие параметры для Random Forest: {random_search_rf.best_params_}")

# Предсказание и оценка лучшей модели
rf_best_preds = best_rf.predict(X_test)

mae_rf_best = mean_absolute_error(y_test, rf_best_preds)
rmse_rf_best = np.sqrt(mean_squared_error(y_test, rf_best_preds))
r2_rf_best = r2_score(y_test, rf_best_preds)

print(f"Best Random Forest - MAE: {mae_rf_best:.2f}, RMSE: {rmse_rf_best:.2f}, R^2: {r2_rf_best:.2f}")
```

```python


# Важность признаков для Random Forest
importances_rf = best_rf.feature_importances_
indices_rf = np.argsort(importances_rf)[::-1]

plt.figure(figsize=(12,8))
plt.title("Важность признаков Random Forest")
plt.bar(range(X_train.shape[1]), importances_rf[indices_rf], align="center")
plt.xticks(range(X_train.shape[1]), X_train.columns[indices_rf], rotation=90)
plt.tight_layout()
plt.show()
```
